```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Mall_Customers.csv")
data
```

```python
df=data.copy()
```

```python
df=pd.DataFrame(data)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.replace(r'^\s*$',pd.NA,regex=True,inplace=True)
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
placeholders=["Unknown","N/A","None","-","?"]
df.replace(placeholders,pd.NA,inplace=True)
print(df.isnull().sum())
```

```python
print(df[df['Annual Income (k$)']==0])
print(df[df['Spending Score (1-100)']==0])
```

```python
df.dtypes
```

```python
print(df.isnull().sum())
```

```python
sns.countplot(data=df,x="Gender",palette='pastel')
plt.title('Number of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Age',data=df,palette='coolwarm')
plt.title("Age Distribution of Mall Customers")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
df.to_csv("cleaned_mallcustomers_data.csv",index=False)
print("Cleaned dataset saved as cleaned_mallcustomers_data.csv")
```
